```python
import pandas as pd
```

```python
aisles = pd.read_csv("data/aisles.csv")
departments = pd.read_csv("data/departments.csv")
order_products_prior = pd.read_csv("data/order_products__prior.csv")
order_products_train = pd.read_csv("data/order_products__train.csv")
orders = pd.read_csv("data/orders.csv")
products = pd.read_csv("data/products.csv")
```

```python
aisles.head()
departments.head()
order_products_prior.head()
orders.head()
products.head()
```

```python
order_products_prior_sample = order_products_prior.sample(200000)
```

```python
df = order_products_prior_sample.merge(products, on="product_id", how="left")
df = df.merge(aisles, on="aisle_id", how="left")
df = df.merge(departments, on="department_id", how="left")

df.head()
```

```python
top_produtos = (
    df.groupby("product_name")["add_to_cart_order"]
      .count()
      .sort_values(ascending=False)
      .head(10)
)

top_produtos
```

```python
top_departamentos = (
    df.groupby("department")["add_to_cart_order"]
      .count()
      .sort_values(ascending=False)
      .head(10)
)

top_departamentos
```

```python
orders_sample = orders.sample(200000)

df_orders = orders_sample.merge(order_products_prior_sample, on="order_id", how="inner")

df_orders["order_hour_of_day"].value_counts().sort_index()
```

```python
top_produtos_df = top_produtos.reset_index()
top_produtos_df.columns = ["produto", "qtd_vendida"]
top_produtos_df
```

```python
top_departamentos_df = top_departamentos.reset_index()
top_departamentos_df.columns = ["departamento", "qtd_vendida"]
top_departamentos_df
```

```python
# Contagem de pedidos por hora do dia
pedidos_por_hora = orders["order_hour_of_day"].value_counts().sort_index()

pedidos_por_hora_df = pedidos_por_hora.reset_index()
pedidos_por_hora_df.columns = ["hora_do_dia", "qtd_pedidos"]

pedidos_por_hora_df
```

```python
# Contagem de pedidos por dia da semana
pedidos_por_dia = orders["order_dow"].value_counts().sort_index()

pedidos_por_dia_df = pedidos_por_dia.reset_index()
pedidos_por_dia_df.columns = ["dia_da_semana", "qtd_pedidos"]

pedidos_por_dia_df
```

```python
# Conta quantos produtos h√° para cada order_id
itens_por_pedido = order_products_train.groupby("order_id")["product_id"].count()

media_itens = itens_por_pedido.mean()
media_itens
```

```python
import pandas as pd

aisles = pd.read_csv("data/aisles.csv")
departments = pd.read_csv("data/departments.csv")
order_products_prior = pd.read_csv("data/order_products__prior.csv")
order_products_train = pd.read_csv("data/order_products__train.csv")
orders = pd.read_csv("data/orders.csv")
products = pd.read_csv("data/products.csv")
```

```python
# Filtra linhas onde o produto foi reordenado
reordenados = order_products_prior[order_products_prior["reordered"] == 1]

# Conta quantas vezes cada produto foi reordenado
top_reordenados = reordenados["product_id"].value_counts().head(10)

# Transformar em DataFrame
top_reordenados_df = top_reordenados.reset_index()
top_reordenados_df.columns = ["product_id", "qtd_reordenado"]

# Juntar com nomes dos produtos
top_reordenados_df = top_reordenados_df.merge(products, on="product_id", how="left")

# Resultado final
display(top_reordenados_df[["product_name", "qtd_reordenado"]])
```

```python
# Salvar top produtos reordenados
top_reordenados_df[["product_name", "qtd_reordenado"]].to_csv("top10_reordenados.csv", index=False)
```
